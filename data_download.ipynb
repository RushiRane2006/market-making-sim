```python
# import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# data = yf.download("AAPL", period="5d", interval="1m")
# data.to_csv("apple_data.csv") 

df = pd.read_csv("apple_data.csv", index_col=0, parse_dates=True)

# df['Close'].plot(title="Apple Stock Price (1m Interval)")
# plt.figure(figsize=(12, 6))
# plt.xlabel("Time")
# plt.ylabel("Price")
# plt.show()

plt.style.use('fivethirtyeight') 

plt.figure(figsize=(12, 6))

plt.plot(df.index, df['Close'], label='AAPL Close', color='#1f77b4', linewidth=1.5)
df['SMA_20'] = df['Close'].rolling(window=20).mean()
plt.plot(df.index, df['SMA_20'], label='20m Moving Avg', color='#ff7f0e', linestyle='--')

plt.title('Apple Inc. - Intraday Price Analysis', fontsize=14, fontweight='bold')
plt.xlabel('Date & Time', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
